```python
threshold = 0.45
null_threshold = len(df) * threshold
df = df.dropna(thresh=null_threshold, axis=1)
print(df.shape)
```

```python
nan_percentage = df.isnull().mean() * 100
print("\nPercentage of NaN values in each feature:")
print(nan_percentage)
```

```python
date_columns = [
    'last_milestone_at', 'first_milestone_at', 'founded_at'
]

for col in date_columns:
  df[col] = pd.to_datetime(df[col], format='%Y-%m-%d')
```

```python
df['founded_year'] = df['founded_at'].dt.year
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 0.1','name','permalink','homepage_url','logo_url'], inplace=True)
df.drop(columns=['normalized_name'], inplace=True)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col].fillna('Unknown', inplace=True)
```

```python
numerical_columns = df.select_dtypes(include=['number']).columns
for col in numerical_columns:
    df[col].fillna(0, inplace=True)
```

```python
date_columns = ['founded_at', 'closed_at', 'first_funding_at', 'last_funding_at', 'first_milestone_at', 'last_milestone_at']

existing_date_columns = [col for col in date_columns if col in df.columns]

for col in existing_date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python

remaining_nan_percentage = df.isnull().mean() * 100
print("\nRemaining Percentage of NaN values after cleaning:")
print(remaining_nan_percentage[remaining_nan_percentage > 0])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

high_cardinality_columns = [col for col in df.select_dtypes(include=['object']).columns if df[col].nunique() > 100]


for col in high_cardinality_columns:
    df[col] = le.fit_transform(df[col].astype(str))


low_cardinality_columns = [col for col in df.select_dtypes(include=['object']).columns if col not in high_cardinality_columns]
df_encoded = pd.get_dummies(df, columns=low_cardinality_columns, drop_first=True)
```

```python

if 'status' in df_encoded.columns:
   
    df_encoded['status_label'] = df_encoded['status'].map({'acquired': 1, 'operating': 0})
    print("Status column successfully encoded.")
else:
    print("The 'status' column is not present in the DataFrame.")
```

```python
print("\nProcessed Data:")
print(df_encoded.head())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
num_features = len(df.columns)
num_rows = (num_features // 4) + (1 if num_features % 4 != 0 else 0)

fig, axs = plt.subplots(num_rows, 4, figsize=(50, num_rows * 7.5))
count = 0
cols = df.columns.to_list()

for i in range(num_rows):
    for j in range(4):
        if count < num_features:
            sns.histplot(df[cols[count]], ax=axs[i, j], kde=True)
            axs[i, j].set_title(cols[count], fontsize=30)
            count += 1
        else:
            axs[i, j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

if 'status' in df.columns:
    df_encoded['status_label'] = df['status'].map({'acquired': 1, 'operating': 0, 'closed': 2, 'ipo': 3})
    print("Status column successfully encoded into 'status_label'.")
else:
    print("The 'status' column is not present in the DataFrame.")

if 'status_label' not in df_encoded.columns:
    raise KeyError("The 'status_label' column was not created or was dropped.")


corr_matrix = df_encoded.corr()


if 'status_label' in corr_matrix.columns:
    corr = np.array(corr_matrix['status_label'])
else:
    raise KeyError("The 'status_label' column is not in the correlation matrix. Ensure it is created and retained properly.")

cols = df_encoded.columns
columns_to_drop = []


for i in range(len(cols)):
    print(cols[i], "-->", np.round(corr[i], 3))
    if np.isnan(corr[i]) or np.abs(corr[i]) < 0.009:
        columns_to_drop.append(cols[i])


essential_columns = ['status_label', 'status_closed', 'status_ipo', 'status_operating']
columns_to_drop = [col for col in columns_to_drop if col not in essential_columns]

df_encoded.drop(columns_to_drop, axis=1, inplace=True)


X = df_encoded.drop(columns=['status_label']).values
y = df_encoded['status_label'].values

print("Columns dropped due to low correlation:", columns_to_drop)
print("Processed Data:")
print(df_encoded.head())
```
